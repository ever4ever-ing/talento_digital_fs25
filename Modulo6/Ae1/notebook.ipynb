```python
import numpy as np

import matplotlib.pyplot as plt

# Datos simulados para población joven y tasas de matrimonio hasta 2030
anios = np.arange(2020, 2031)
poblacion_joven = np.linspace(10_000_000, 9_000_000, len(anios))  # Ejemplo: población decreciente
tasa_matrimonio = np.linspace(30, 20, len(anios))  # Ejemplo: tasa decreciente (%)

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Año')
ax1.set_ylabel('Población joven', color=color)
ax1.plot(anios, poblacion_joven, color=color, label='Población joven')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Tasa de matrimonio (%)', color=color)
ax2.plot(anios, tasa_matrimonio, color=color, linestyle='--', label='Tasa de matrimonio')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Proyección de población joven y tasas de matrimonio hacia 2030')
fig.tight_layout()
plt.show()
```
